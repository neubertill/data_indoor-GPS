```python
import pandas as pd #for working with dataframes
import numpy as np #for numerical and vector operations
import matplotlib as mpl #for visualization
import matplotlib.pyplot as plt #for visualization
import seaborn as sns #for visualization
import os
from adjustText import adjust_text
```

```python
df1=pd.read_csv("Scenario1.csv", sep= ';',header=0)
print(df1)
xR=np.asarray(df1["xR"])
yR=np.asarray(df1["yR"])

xL=np.asarray(df1["xL"])
yL=np.asarray(df1["yL"])

nR=df1["Reference"]
nL=df1["Localino"]
```

```python
fig, ax = plt.subplots(figsize=(18,10))
ref=plt.scatter(xR, yR,c='tab:orange')
loc=plt.scatter(xL, yL,c='tab:blue')
plt.grid(True)
plt.xlabel("x [cm]")
plt.ylabel("y [cm]")
plt.xticks([-250,-200,-150,-100,-50,0,50,100,150,200,250])


circle1 = mpl.patches.Ellipse((229, 63), 11*2,8*2, color='r',fill=False)
circle2 = mpl.patches.Ellipse((187, 52), 11*2,8*2, color='r',fill=False)
circle3 = mpl.patches.Ellipse((134, 46), 11*2,8*2, color='r',fill=False)
ax.add_patch(circle1)
ax.add_patch(circle2)
ax.add_patch(circle3)

ax.legend([ref,loc,circle1],['Laser tracker','Indoor GPS','Outlier'],title='Scenario 1')

texts = [plt.text(xR[i], yR[i], nR[i], ha='center', va='center') for i in range(len(xR))]
texts2 = [plt.text(xL[i], yL[i], nL[i], ha='center', va='center') for i in range(len(xL))]
adjust_text(texts+texts2, avoid_self=False)

plt.savefig("Scenario1_fig.png")
```
